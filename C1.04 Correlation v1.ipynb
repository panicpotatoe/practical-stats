```python
import pandas as pd
SP500_SECTORS_CSV = "https://raw.githubusercontent.com/panicpotatoe/dataset-practical-stats/main/sp500_sectors.csv"
SP500_DATA_CSV = "https://raw.githubusercontent.com/panicpotatoe/dataset-practical-stats/main/sp500_data.csv.gz"
```

```python
sp500_sym = pd.read_csv(SP500_SECTORS_CSV)
sp500_sym.head()
```

```python
sp500_px = pd.read_csv(SP500_DATA_CSV, index_col=0)
sp500_px.head()
```

```python
# Determine telecommunications symbols
telecomSymbols = sp500_sym[sp500_sym['sector'] == 'telecommunications_services']['symbol']
telecomSymbols
```

```python
# Filter data for dates July 2012 through June 2015
telecom = sp500_px.loc[sp500_px.index >= '2012-07-01', telecomSymbols]
telecom
```

```python
telecom.corr()
print(telecom)
```

```python
# (temp) sp500_px after 01 July 2012
df1 = sp500_px[sp500_px.index > '2012-07-01']
df1.head()
```

```python
# (temp) sp500_sym that is etf
# etf = exchange traded fund
df2 = sp500_sym[sp500_sym['sector'] == 'etf']
df2.head()
```

```python
# (temp) sp500_sym that is etf + get its symbol only
df3 = sp500_sym[sp500_sym['sector'] == 'etf']['symbol']
df3.head()
```

```python
# (final) sp500_px after 01 July 2012 + filter symbols that are in etf sector only
etfs = sp500_px.loc[sp500_px.index > '2012-07-01', 
                    sp500_sym[sp500_sym['sector'] == 'etf']['symbol']]
etfs.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(figsize=(5, 4))
ax = sns.heatmap(etfs.corr(), vmin=-1, vmax=1, 
                 cmap=sns.diverging_palette(20, 220, as_cmap=True),
                 ax=ax)

plt.tight_layout()
plt.show()
```
