```python
import pandas as pd
state = pd.read_csv('https://raw.githubusercontent.com/panicpotatoe/dataset-practical-stats/main/state.csv')
```

```python
import matplotlib.pyplot as plt
ax = (state['Population'] / 1_000_000).plot.hist(figsize=(4, 4))
ax.set_xlabel('Population (millions)')

plt.tight_layout()
plt.show()
```

```python
ax = state['Murder.Rate'].plot.hist(density=True, xlim=[0, 12],
                                    bins=range(1,12), figsize=(4, 4))
state['Murder.Rate'].plot.density(ax=ax)
ax.set_xlabel('Murder Rate (per 100,000)')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/panicpotatoe/dataset/master/SOCR_Data_Dinov_020108_HeightsWeights.csv")
df.sample()
```

```python
import matplotlib.pyplot as plt
plt.hist(df["Height(Inches)"], bins=50)
plt.show()
```

```python
import numpy as np
mean_h = np.mean(df["Height(Inches)"])
median_h = np.median(df["Height(Inches)"])
print(f'{mean_h=}')
print(f'{median_h=}')
```

```python
plt.hist(df["Height(Inches)"], bins=100)
plt.axvline(mean_h, color='r', linestyle='dashed', linewidth=1)
plt.axvline(median_h, color='orange', linestyle='dashed', linewidth=1)
plt.show()
```

```python
mean_w = np.mean(df["Weight(Pounds)"])
median_w = np.median(df["Weight(Pounds)"])
print(f'{mean_w=}')
print(f'{median_w=}')
```

```python
plt.figure(figsize=(20, 5))
plt.hist(df["Weight(Pounds)"], bins=100)
plt.axvline(mean_w, color='r', linestyle='dashed', linewidth=1)
plt.axvline(median_w, color='orange', linestyle='dashed', linewidth=1)
plt.show()
```

```python
plt.figure(figsize=(20, 2))
plt.boxplot(df["Weight(Pounds)"], vert=False)
plt.show()
```
