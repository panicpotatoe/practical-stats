```python
import pandas as pd
AIRPORT_DELAYS_CSV = "https://raw.githubusercontent.com/panicpotatoe/dataset-practical-stats/main/dfw_airline.csv"
dfw = pd.read_csv(AIRPORT_DELAYS_CSV)
```

```python
dfw
```

```python
# dfw in 100 percent
print(100 * dfw / dfw.values.sum())
```

```python
# barplot of dfw in 100 percent
# approach 1
dfw.plot.bar()
```

```python
# approach 2
import matplotlib.pyplot as plt
ax = dfw.transpose().plot.bar(figsize=(4, 4), legend=False)
ax.set_xlabel('Cause of delay')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```
